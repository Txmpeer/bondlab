```python
import bondlab as bl
```

## 1. Discount Factors

We start by computing discount factors under different compounding conventions.

```python
df_discrete = bl.discount_factor(0.05, 1, frequency=2, compounding="discrete")
df_continuous = bl.discount_factor(0.05, 1, compounding="continuous")

print("Discrete:", df_discrete)
print("Continuous:", df_continuous)
```

## 2. Zero-Coupon Bond Pricing

```python
zc_price = bl.price_zero_coupon_bond(
    face=100,
    rate=0.05,
    maturity=2,
    frequency=2
)

print("Zero-coupon price:", zc_price)
```

## 3. Coupon Bond Cash Flows and Pricing

```python
times, cashflows = bl.bond_cashflows(
    face=100,
    coupon_rate=0.06,
    maturity=2,
    frequency=2
)

print("Times:", times)
print("Cashflows:", cashflows)

price = bl.price_coupon_bond_from_ytm(
    face=100,
    coupon_rate=0.06,
    ytm=0.05,
    maturity=2,
    frequency=2
)

print("Coupon bond price:", price)
```

## 4. Yield to Maturity

```python
ytm = bl.yield_to_maturity(
    price=price,
    face=100,
    coupon_rate=0.06,
    maturity=2,
    frequency=2
)

print("Recovered YTM:", ytm)
```

## 5. Duration and Convexity

```python
mac = bl.macaulay_duration(
    face=100,
    coupon_rate=0.06,
    maturity=2,
    rate=0.05,
    frequency=2
)

mod = bl.modified_duration(
    face=100,
    coupon_rate=0.06,
    maturity=2,
    rate=0.05,
    frequency=2
)

cx = bl.convexity(
    face=100,
    coupon_rate=0.06,
    maturity=2,
    rate=0.05,
    frequency=2
)

print("Macaulay:", mac)
print("Modified:", mod)
print("Convexity:", cx)
```

## 6. Spot Rate Conversions and Forward Rates

```python
df = bl.spot_to_discount_factor(0.05, 2)
spot = bl.discount_factor_to_spot(df, 2)
fwd = bl.forward_rate(0.04, 1, 0.05, 2, compounding="continuous")

print("DF:", df)
print("Spot:", spot)
print("Forward:", fwd)
```

## 7. Interpolation and Bootstrapping

```python
interp = bl.interpolate_rate([1, 2], [0.04, 0.06], 1.5)
print("Interpolated:", interp)

prices = [
    100 * bl.spot_to_discount_factor(0.05, 1, frequency=1),
    100 * bl.spot_to_discount_factor(0.06, 2, frequency=1),
]

curve = bl.bootstrap_spot_curve(
    prices=prices,
    maturities=[1, 2],
    coupon_rates=[0.0, 0.0],
    face=100,
    frequency=1
)

print(curve)
```

## 8. Pricing from a Spot Curve

```python
curve_price = bl.price_bond_from_spot_curve(
    face=100,
    coupon_rate=0.05,
    maturity=3,
    spot_rates=[0.04, 0.045, 0.05],
    times=[1, 2, 3],
    frequency=1
)

print("Price from curve:", curve_price)
```

## 9. Portfolio Analytics

```python
portfolio_dur = bl.portfolio_duration([0.4, 0.6], [4.0, 6.0])
portfolio_cx = bl.portfolio_convexity([0.4, 0.6], [10.0, 14.0])

print("Portfolio duration:", portfolio_dur)
print("Portfolio convexity:", portfolio_cx)
```

## 10. Factor-Based Sensitivities

```python
level = bl.level_duration_from_curve(
    face=100,
    coupon_rate=0.05,
    maturity=3,
    spot_rates=[0.04, 0.045, 0.05],
    times=[1, 2, 3],
    frequency=1
)

slope = bl.slope_duration_from_curve(
    face=100,
    coupon_rate=0.05,
    maturity=3,
    spot_rates=[0.04, 0.045, 0.05],
    times=[1, 2, 3],
    frequency=1
)

print("Level duration:", level)
print("Slope duration:", slope)
```
